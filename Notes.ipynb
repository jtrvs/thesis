## Глубина покрытия для SNP и инделей полногенома
* минимум 10х, оптимум 15-30, для популяционных исследований достаточно 15
* для структурных вариантов 50-100

* QD (Quality by Depth): > 2.0 (качество варианта, нормированное на глубину).
* GQ (Genotype Quality): > 20 для генотипов.
* AD (Allelic Depth): Минимум 3–5 прочтений на альтернативный аллель

* DP < 10 — высокий риск ложных вызовов вариантов.
* DP > 100 — возможны артефакты ПЦР-дуплика

### Когда что использовать касательно фильтрации по DP?
aggregate_entries

* Поиск систематических проблем (например, образцы с низким DP).

* Анализ качества вызова вариантов.
* Фильтрация редких артефактов (например, варианты с mean_DP < 10).
